```python
import xml.etree.ElementTree as ET
tree = ET.parse('sample.xml')
root = tree.getroot()

# all items data

for elem in root:
    print(elem.tag)
    print(elem.text)
    
    print("--subtag--")
    
    for subelem in elem:
        print(subelem.tag)
        print(subelem.text)
        
        if subelem.tag == "bndbox":
            for subsubelem in subelem:
                print(subsubelem.tag)
                print(subsubelem.text)

    print("----")
```
