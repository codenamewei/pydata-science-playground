```python
import pandas as pd 
import numpy as np

rows = 5
df = pd.DataFrame(np.random.randint(0,rows,size=(rows, 2)), columns=list('AB'))

df["C"] = [0,0, 1, 1, 0]
df["D"] = [-1, 6, -1, -1, 6]

df.head(5)
```

### Assign value of B = 1000 where C = 1 & D = -1 &

```python
df2 = df
df2.loc[df2.D.eq(-1), 'B'] = 1000

df2.head(5)
```

```python
df3 = df
df3.loc[df3.D.eq(-1) & df3.C.eq(1), 'B'] = 1000

df3.head(5)
```

### this works, but with warnings

```python
df1 = df
df1[df1.D == -1].B = 1000
df1.head(5)
```
